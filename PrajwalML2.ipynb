```python
import pandas as pd 
import matplotlib.pyplot as plt 
import math
```

1. Read the temperature data

```python
df=pd.read_csv("/Users/prajwal7/Desktop/PRAJWAL/CODING/TE ML/temperatures.csv")
```

```python
df.head()
```

2. Check if data has null or duplicated values

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

#  1) February Month Prediction

```python
X=df[['YEAR']]
Y=df[['FEB']]
```

3. Visualize the Whole Dataset

```python
plt.scatter(X,Y)
```

4. Split the dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
plt.scatter(X_train,Y_train)
```

5. Train the model on Training data

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

6. Make Predictions on test data

```python
Y_pred=model.predict(X_test)
```

7. Find Y intercept and Coefficient

```python
print(f"Y Intercept={model.intercept_}")
print(f"Coefficient={model.coef_}")
```

8. Visualize the model with Training data

```python
plt.xlabel("YEAR")
plt.ylabel("TEMPERATURE")
plt.title("  TEMPERATURE VS YEAR")
plt.scatter(X_train,Y_train,color="red")
plt.plot(X_test,Y_pred)
plt.show()
```

9. Find the Performance using MAE, MSE and RMSE

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(f"MAE={mean_absolute_error(Y_test,Y_pred)}")
print(f"MSE={mean_squared_error(Y_test,Y_pred)}")
print(f"RMSE={math.sqrt(mean_squared_error(Y_test,Y_pred))}")
```
